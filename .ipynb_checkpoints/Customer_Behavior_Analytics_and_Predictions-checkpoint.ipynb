```python
# 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)
```

```python
# 2. Loading the dataset
file_path = 'Online_Retail_Customer_Transactions.xlsx'
df = pd.read_excel(file_path)
```

```python
# 3. Data Cleaning and Preparation
# Checking for null values
print(df.isnull().sum())

# The dataset is pre-cleaned, so we expect no missing values.
# However, if there are any, we can choose to drop or impute them.

# Ensuring correct data types
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['ReturnStatus'] = df['ReturnStatus'].astype(int)  # Convert boolean to int

# Removing any duplicate entries if they exist
df.drop_duplicates(inplace=True)
```

# Data Cleaning and Preparation

## Null Value Check
The initial check for null values across all columns has returned zeros for all fields, indicating that there are no missing values in any column. This suggests that the dataset is well-maintained and does not require any imputation or deletion of records due to missing data, which is an excellent starting point for any data analysis project.

## Data Type Confirmation
The data types for `TransactionDate` and `ReturnStatus` have been confirmed and adjusted as necessary. The `TransactionDate` has been converted to a datetime object, which is essential for any time-series analysis or operations that require date arithmetic. The `ReturnStatus` column has been converted from boolean to integer format. This conversion is likely intended to prepare the dataset for machine learning models, as many algorithms expect numerical input.

## Duplicate Entries
The dataset has been checked for duplicate entries, and none have been found. This step is vital because duplicates can lead to skewed results and analyses that may misrepresent the true nature of the data.

The results from this phase assure us that the dataset is in a robust state for the next steps of the project. With a clean dataset, we can proceed confidently, knowing that the foundation of our analysis is solid.


```python
# 4. Exploratory Data Analysis (EDA)

# Summary statistics
print(df.describe())

# Visualizing distributions of key variables
sns.histplot(df['TotalAmount'], kde=True)
plt.title('Distribution of Total Amount')
plt.show()

sns.countplot(x='CustomerRating', data=df)
plt.title('Distribution of Customer Ratings')
plt.show()

# Relationship between TotalAmount and CustomerRating
sns.scatterplot(x='TotalAmount', y='CustomerRating', data=df)
plt.title('Total Amount vs Customer Rating')
plt.show()
```

```python
# 5. Feature Engineering

# Creating a feature for average spending per product
df['AvgSpendingPerProduct'] = df['TotalAmount'] / df['Quantity']
```

```python
# 6. Customer Segmentation using K-Means

# Selecting features for clustering
features = df[['TotalAmount', 'CustomerRating', 'AvgSpendingPerProduct']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying K-Means
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)# Number of clusters chosen based on project needs
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Visualizing the clusters
sns.scatterplot(data=df, x='TotalAmount', y='CustomerRating', hue='Cluster', palette='viridis')
plt.title('Customer Segments')
plt.show()
```

# Customer Segmentation using K-Means

The K-Means algorithm was applied to the dataset to segment customers based on three key features: `TotalAmount`, `CustomerRating`, and `AvgSpendingPerProduct`. The selection of these features aims to capture different aspects of customer behavior: total spending, satisfaction, and spending per product, which could indicate the value customers place on the products they purchase.

## K-Means Clustering

- The dataset was scaled using `StandardScaler` to ensure that each feature contributes equally to the distance calculations during clustering.
- K-Means was initialized with 4 clusters (`n_clusters=4`), and the algorithm was run 10 times (`n_init=10`) with different centroid seeds to ensure the robustness of the solution.
- The `random_state=42` was set for reproducibility of results.

## Visualization of Clusters

- The scatter plot shows the distribution of customers across the 4 clusters based on `TotalAmount` and `CustomerRating`, with different colors representing different clusters.
- It appears that the clusters do not show a distinct separation based on `TotalAmount` alone, which could suggest that the customer segments are more nuanced and possibly influenced by the combination of features used.

## Insights

- The clustering resulted in four distinct customer groups, with no clear separation along the `TotalAmount` dimension. This could indicate that `TotalAmount` isn't the primary driver of differentiation among the clusters.
- A possible next step could be to analyze the clusters in more detail, looking at the centroids of the clusters for insights into the defining characteristics of each segment, or consider additional features that could further differentiate the customer groups.
- The segments could be profiled based on the clustering to understand the average spending per product and rating behavior within each cluster.

The K-Means clustering has provided a starting point for understanding the different types of customers in the dataset. The insights gained from this segmentation can inform targeted marketing strategies, product placement, and customer service initiatives.


```python
#7. Predictive Modeling (Predicting ReturnStatus)

# Preparing data for modeling
X = df[['TotalAmount', 'CustomerRating', 'Quantity', 'Cluster']]
y = df['ReturnStatus']  # Already converted to int in data cleaning step

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Building the RandomForest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Making predictions and evaluating the model
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

# Predictive Modeling using Random Forest

In this phase, we employed a Random Forest Classifier to predict the `ReturnStatus` of transactions. Random Forest is an ensemble learning method known for its high accuracy and robustness, as it combines the predictions of multiple decision trees to improve predictive performance and control over-fitting.

## Model Preparation and Training
- The features chosen for the model included `TotalAmount`, `CustomerRating`, `Quantity`, and the cluster labels obtained from K-Means segmentation. This combination captures both the spending behavior and the customer satisfaction metrics.
- The data was split into a training set and a test set, with 70% of the data used for training and 30% used for testing. This split allows us to assess the model's generalization on unseen data.
- The model was trained on the training dataset with a specified `random_state` to ensure that the results are repeatable.

## Model Evaluation
- After training, the model was used to predict the `ReturnStatus` for the test set.
- A classification report was generated, providing a detailed account of the model's precision, recall, and F1-score for each class.
- Additionally, a confusion matrix was created to visually assess the model's performance, showing the number of true positives, false positives, false negatives, and true negatives.

## Results and Interpretation
- The classification report indicates an almost even precision and recall for both classes, suggesting the model does not heavily favor one class over the other.
- The accuracy is around 48%, which suggests the model is only slightly better than a random guess. This may be due to the inherent complexity of the problem or limitations in the feature set used for modeling.
- The confusion matrix shows a relatively balanced distribution of predictions across the two classes but also highlights a considerable number of misclassifications.

## Conclusion
The predictive modeling has yielded a baseline performance from which we can iterate. The moderate accuracy indicates there is room for improvement. Future steps could involve more sophisticated feature engineering, hyperparameter tuning, or trying different machine learning algorithms to improve the model's predictive ability. The ultimate goal is to develop a model that can accurately predict customer returns, which is crucial for managing inventory and customer satisfaction.
